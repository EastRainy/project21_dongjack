```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
df = pd.read_csv('../data/housing.csv', delim_whitespace=True, header=None)
df.info()
```

```python
df.corr()
```

```python
seed = 1234
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
dataset = df.values
X = dataset[:, :13]
Y = dataset[:, 13]

print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=seed)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=13, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['MAE'])
```

```python
model.fit(X_train, Y_train, epochs=500, batch_size=20)
```

```python
result = model.evaluate(X_test, Y_test, verbose=0)

print('test MAE:', result[1])
```

```python
Y_pred = model.predict(X_test)
```

```python
print(X_test.shape, Y_pred.shape, Y_test.shape)
```

```python
Y_pred = model.predict(X_test).flatten()
```

```python
print(X_test.shape, Y_pred.shape, Y_test.shape)
```

```python
for y_pred, y in zip(Y_pred, Y_test):
    print(' %5.2f, : %5.2f, : %06.2f' % (y_pred, y, y - y_pred))
```
