```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = np.loadtxt('../data/ThoraricSurgery.csv', delimiter=',')

print(type(dataset), dataset.shape, dataset.dtype)
```

```python
X = dataset[:, :17]
Y = dataset[:, 17]

print(type(X), X.shape, X.dtype)
print(type(Y), Y.shape, Y.dtype)
```

```python
np.random.seed(1234)
tf.random.set_seed(1234)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=100, batch_size=10)
```

```python
model = Sequential()
model.add(Dense(32, input_dim=17, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=200, batch_size=50)
```
