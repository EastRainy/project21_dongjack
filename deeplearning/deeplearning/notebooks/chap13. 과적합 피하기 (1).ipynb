```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
```

```python
tf.__version__
```

```python
df = pd.read_csv('../data/sonar.csv', header=None)
df.head()
```

```python
df.info()
```

```python
'''
문제] df 데이터프레임의 데이터를 이용하여 학습 가능한 훈련 데이터 X, Y 를 생성하는 코드를 작성하세요.
'''
```

```python
dataset = df.values
X = dataset[:, :60].astype(np.float64)
Y_obj = dataset[:, 60]

print(X.shape, X.dtype, Y_obj.shape, Y_obj.dtype)
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

print(Y.shape, Y.dtype)
Y
```

```python
seed = 1234
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=200, batch_size=5)
```

```python
result = model.evaluate(X, Y, verbose=0)
print('loss: {:.4f}, accuracy: {:.4f}'.format(result[0], result[1]))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=5)
```

```python
result = model.evaluate(X_test, Y_test, verbose=0)

print('test loss: %.4f, test accuracy: %.4f' % (result[0], result[1]))
```
