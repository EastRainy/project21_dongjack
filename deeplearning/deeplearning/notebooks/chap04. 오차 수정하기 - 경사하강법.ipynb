```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
X = np.array([2, 4, 6, 8])
Y = np.array([81, 93, 91, 97])
```

```python
plt.scatter(X, Y)
plt.grid()

plt.show()
```

```python
a, b = 0, 0
lr = 0.03
epochs = 2001
```

```python
for i in range(epochs):
    Y_pred = a * X + b
    error = Y - Y_pred
    
    a_diff = -(2 / len(X)) * (X * error).sum()
    b_diff = -(2 / len(X)) * error.sum()
    
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print('epoch: %4d. a: %.4f, b: %.4f' % (i, a, b))
        
print('a: %.4f, b: %.4f' % (a, b))
```

```python
Y_pred = a * X + b

plt.scatter(X, Y)
plt.scatter(X, Y_pred)
plt.plot(X, Y_pred)
plt.grid()

plt.show()
```

```python
X1 = np.array([2, 4 ,6 ,8])
X2 = np.array([0, 4, 2, 3])
Y = np.array([81, 93, 91, 97])
```

```python
ax = plt.axes(projection='3d')
ax.scatter(X1, X2, Y)

ax.set_xlabel('Study hours')
ax.set_ylabel('Private_class')
ax.set_zlabel('Score')
ax.dist = 11
plt.yticks([0, 1, 2, 3, 4])
plt.show()
```

```python
a1, a2, b = 0, 0, 0
lr = 0.02
epochs = 2001
```

```python
for i in range(epochs):
    Y_pred = a1 * X1 + a2 * X2 + b
    error = Y - Y_pred
    
    a1_diff = -(2 / len(X1)) * (X1 * error).sum()
    a2_diff = -(2 / len(X1)) * (X2 * error).sum()
    b_diff = -(2 / len(X1)) * error.sum()
    
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print('epoch: %4d, a1: %.4f, a2: %.4f, b: %.4f' % (i, a1, a2, b))
```

```python
Y_pred = a1 * X1 + a2 * X2 + b

print('실제값:', Y)
print('예측값:', Y_pred)
```
