```python
url = 'https://www.alexa.com/topsites/countries/KR'

html = requests.get(url).text
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
site_ranking = soup.select('p a')
print(len(site_ranking), site_ranking[0], site_ranking[-1], sep='\n')
```

```python
site_ranking = soup.select('div.tr.site-listing>div.td.DescriptionCell>p>a')
print(len(site_ranking), site_ranking[0], site_ranking[-1], sep='\n')
```

```python
'''
문제] 사이트 순위(1 ~ 50)를 인덱스로 갖고 사이트 주소를 site 컬럼의 값으로 갖는 site_ranking 데이터프레임을
        생성하는 코드를 작성하세요.
'''
```

```python
ranking_list = []

for site in site_ranking:
    ranking_list.append(site.get_text())
    
site_ranking = pd.DataFrame(ranking_list, index=range(1, len(ranking_list) + 1), columns=['site'])
site_ranking
```

```python
url = 'https://www.python.org/static/img/python-logo.png'
image_file_name = os.path.basename(url)
image_file_name
```

```python
folder = '../download/'

if not os.path.exists(folder):
    os.makedirs(folder)
```

```python
save_file_name = folder + image_file_name
save_file_name
```

```python
r = requests.get(url)
r
```

```python
f = open(save_file_name, 'wb')

chunk_size = 1024
for chunk in r.iter_content(chunk_size):
    f.write(chunk)
    
f.close()
```
